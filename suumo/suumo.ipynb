```python
data = []
#それぞれのURLのページを繰り返し実行する
for u in range(1,117) :

    url = f'https://suumo.jp/jj/chintai/ichiran/FR301FC001/?ar=030&bs=040&ta=13&sc=13114&cb=0.0&ct=9999999&et=9999999&cn=9999999&mb=0&mt=9999999&shkr1=03&shkr2=03&shkr3=03&shkr4=03&fw2=&srch_navi=1&page={u}'
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    for i in range(1,30):
        #部屋の情報を取得
        houses = soup.find_all('div',class_="cassetteitem")[i]
        #部屋の価格を取得
        price = houses.find('span',class_ = "cassetteitem_other-emphasis ui-text--bold").text
        #部屋の価格を数値に変換
        price = float(price.replace('万円',''))
        #家までの時間を取得
        time = houses.find_all('div',class_ ="cassetteitem_detail-text")[0].text
        #家までの時間を数値に変換
        time = int (time.split('歩')[1].split('分')[0])
        #データを収納
        datum = {}
        #価格,時間をそれぞれ設定
        datum['price'] = price
        datum['time'] = time
        data.append(datum)
    sleep(1)
        
```

```python
import sqlite3

# SQLiteデータベースに接続
conn = sqlite3.connect('suumo_data.db')

# カーソルを取得
cursor = conn.cursor()

# テーブルを作成
cursor.execute('''
CREATE TABLE IF NOT EXISTS rental_properties (
    id INTEGER PRIMARY KEY AUTOINCREMENT, -- 自動的に増加するID
    price REAL,                           -- 賃料（浮動小数点数）
    time INTEGER                          -- 最寄駅からの時間（分）
)
''')

# データを挿入
for datum in data:
    cursor.execute('''
    INSERT INTO rental_properties (price, time) VALUES (?, ?)
    ''', (datum['price'], datum['time']))

# 変更を保存
conn.commit()

# データベース接続を閉じる
conn.close()

print("データベース 'suumo_data.db' にデータを保存しました")
```

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# SQLiteデータベースからデータを取得する関数
def fetch_data_from_db(db_path):
    # SQLiteデータベースに接続
    conn = sqlite3.connect(db_path)
    
    # データを取得するクエリ
    query = "SELECT price, time FROM rental_properties"
    
    # Pandas DataFrame にデータを読み込む
    df = pd.read_sql_query(query, conn)
    
    # データベース接続を閉じる
    conn.close()
    
    return df

# データベースファイルのパス
db_path = 'suumo_data.db'

# データベースからデータを取得
data = fetch_data_from_db(db_path)

# データの確認
print(data.head())

# 散布図の作成
plt.figure(figsize=(10, 6))
plt.scatter(data['time'], data['price'], alpha=0.7, color='blue')

# タイトルとラベルを設定
plt.title('The relationship between time and rent', fontsize=14)
plt.xlabel('time', fontsize=12, )
plt.ylabel('price', fontsize=12, )

# グリッドを追加
plt.grid(True, linestyle='--', alpha=0.7)

# グラフを表示
plt.show()
```

```python
# 時間ごとに平均家賃を計算
average_prices = data.groupby('time')['price'].mean()

# 棒グラフを作成
plt.figure(figsize=(12, 8))
average_prices.plot(kind='bar', color='skyblue', alpha=0.8)

# 見出し
plt.title('The relationship between time and rent', fontsize=16)
plt.xlabel('time', fontsize=14, )
plt.ylabel('price', fontsize=14)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
